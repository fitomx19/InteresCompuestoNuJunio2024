### Primera prueba: 

Si al mes ingreso 10,000 pesos , y no retiro mi ganancia más lo ya previamente invertido. ¿Cuanto tendre a final de el periodo?


Primero haremos el calculo por un mes simple

```python
Ingreso_por_dia_al_mes = Inversion * (1 + interes_diario_por_peso)**30
Ingreso_por_dia_al_mes
```

Nos da como resultado $121.618 pesos de ganancia por interes

Ahora por los *6 meses* del ejercicio

```python
for i in range(0, 360):
    #El primer mes se empieza con 10,000 y se termina con 10,000 + 30 días de intereses
    final_mes = Inversion * (1 + interes_diario_por_peso)**30
    Inversion = final_mes + 10000
    
    print(f"Mes {i}: {final_mes}")
    print(f"Interes: {final_mes - (10000*(i+1))}")
```

Al finalizar el mes 6 obtendremos $2606.387 pesos de ganancia
